## Merge two dataframe

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
                   'Class':[9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4],
                   'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on='ID')
# Here the common index column is 'ID'
```

```python
df3 = pd.DataFrame({'ID':[1,2,3,5],
                   'Name':['A','B','C','D']})
df3
```

```python
pd.merge(df1,df3,on='ID')
# Here the common index column is 'ID'
```

## How (inner,outer,left,right) for merging two dataframe with uncommon datas

```python
pd.merge(df1,df3,on='ID',how='inner')
```

```python
pd.merge(df1,df3,on='ID',how='outer')
```

```python
pd.merge(df1,df3,on='ID',how='left')
```

```python
pd.merge(df1,df3,on='ID',how='right')
```

## Indicator
### It will indicate the merging operation

```python
pd.merge(df1,df3,on='ID',how='left',indicator=True)
```

## Change index name

```python
pd.merge(df1,df3,left_index=True,right_index=True)
```
