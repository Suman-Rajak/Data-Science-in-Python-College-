##  Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Loading Datasets

```python
datairis = sns.load_dataset("iris")
datatips = sns.load_dataset("tips")
datatitanic = sns.load_dataset("titanic")
datapenguins = sns.load_dataset("penguins")
```

## Lineplot

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=datairis)
plt.show()
```

```python
sns.lineplot(x="total_bill",y="size",hue="sex",style="sex",data=datatips)
plt.show()
```

## Scatter Plot

### With Regression Line

```python
sns.lmplot(x='total_bill',y='tip',data=datatips)
plt.show()
```

### Without Regression Line

```python
sns.lmplot(x='total_bill',y='tip',data=datatips,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=datatips,fit_reg=False,hue='sex')
plt.show()
```

## Barplot

```python
sns.barplot(x='who',y='fare',data=datatitanic)
plt.show()
```

```python
sns.barplot(x='who',y='fare',hue="class",data=datatitanic)
plt.show()
```

### Change Colour theme (palette)

```python
sns.barplot(x='who',y='fare',hue="class",data=datatitanic,palette="Blues")
plt.show()
```

```python
sns.barplot(x='who',y='fare',hue="class",data=datatitanic,estimator=np.median,ci=0) #instead we can use "errorbar = ('ci',0)"
plt.show()
```

## Distplot

```python
sns.set_style('whitegrid')
sns.distplot(datatips['total_bill'],kde=False,color='red',bins=30)
```

## Jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=datatips)
```

### Kde (will show the most density portion)

```python
sns.jointplot(x='total_bill',y='tip',data=datatips,kind='kde')
```

## Random Numbers

```python
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var,name="Numerical Variable")
```

## Histogram

```python
sns.histplot(data=num_var,kde=True) #Kde 
```

```python
sns.histplot(data=datapenguins, x="body_mass_g",kde=True)
```

```python
sns.histplot(data=datapenguins, x="body_mass_g",kde=True,hue="species")
```

```python
sns.histplot(data=datatips,x="size",stat="probability",discrete=True)
```

## Pairplot

```python
sns.pairplot(datatips,hue='sex')
```

## Heatmap

### Preparing the data for heatmap

```python
dataforheat = np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(dataforheat)
```

### Plotting the heatmap

```python
hm = sns.heatmap(data=dataforheat)
```

### Changing the minimum and maximum value

```python
minvalue = 30
maxvalue = 70

hm1 = sns.heatmap(data = dataforheat, vmin=minvalue, vmax=maxvalue)
```

### annot

```python
hm1 = sns.heatmap(data = dataforheat, annot=True)
```
