## Statistical Measures of any column

### mean value using only index

```python
df.pivot_table(index='Genre')
```

### mean value using index and columns

```python
df.pivot_table(index='Genre',columns='Year')
```

### Aggregate Function

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='count') #we can use max,min,mean anything using the aggfunc
```

### Fill Any Null Value

```python
df.pivot_table(index='Genre',columns='Year',fill_value=0)
```

### Margins

```python
df.pivot_table(index='Genre',columns='Year',fill_value=0,margins=True)
```

## Reshape a Dataframe using 'melt()'

```python
pd.melt(df)
```

### id_vars (Reshaping by a column)

```python
pd.melt(df,id_vars=['Genre'])
```

### value_vars (Only one value will be there)

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'])
```

### change value name and var name

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'],var_name='Category')
```

```python
pd.melt(df,id_vars=['Year'],value_vars=['Genre'],var_name='Category',value_name='Data')
```
